****1.**** Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
import math as mf
```

```python
# Используем формулу Бернулли, поскольку случайная величина дискристная и вероятность наступления события > 0.1
n = 100
k = 85
p_1 = 0.8

c = mf.comb(n, k)
p = pow(p_1, k)
q = pow(1-p_1, n-k)
print('Вероятность того, что стрелок попадет ровно 85 раз = ', c*p*q)
```

****2.**** Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

```python
# Используем формулу Пуассона, так как случайная величина дискретная и вероятность наступления события < 0.1
n = 5000
p_1 = 0.0004
l = n * p_1

m = 0
p = mf.pow(l, m) / mf.factorial(m) * mf.exp(-1*l)
print('Вероятность того, что ни одна лампочка не перегорит = ', p)

m = 2
p = mf.pow(l, m) / mf.factorial(m) * mf.exp(-1*l)
print('Вероятность того, что перегорят ровно 2 = ', p)
```

****3.**** Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз

```python
# Используем формулу Бернулли, поскольку случайная величина дискристная и вероятность наступления события > 0.1
n = 144
k = 70
p_1 = 0.5

c = mf.comb(n, k)
p = pow(p_1, k)
q = pow(1-p_1, n-k)
print(f'Вероятность того, что орел выпадет {k} раз = {c*p*q}')
```

****4.****
В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

```python
# вероятность того, что из первого ящика выбрали 2 белых мяча
p1_2 = 7/10*6/9

# вероятность того, что из второго ящика выбрали 2 белых мяча
p2_2 = 9/11*8/10

print('Вероятность того, что все мячи белые =', p1_2 * p2_2)
```

```python
# вероятность того, что из первого ящика не выбрали ни одного белого шара (то же, что выбрали оба черных)
p1_0 = 3/10*2/9

# вероятность того, что из первого ящика выбрали один белый и один черный
p1_1_1 = (3/10*7/9) + (7/10*3/9) # сначала черный, потом белый или сначала белый, потом черный

# вероятность того, что из второго ящика не выбрали ни одного белого шара (то же, что выбрали оба черных)
p2_0 = 2/11*1/10

# вероятность того, что из второго ящика выбрали один белый и один черный
p2_1_1 = (2/11*9/10) + (9/11*2/10) # сначала черный, потом белый или сначала белый, потом черный
```

```python
# вероятность того, что из первого ящика выбрали два белых, а из второго два черных
p_ww_bb = p1_2 * p2_0

# вероятность того, что из первого ящика выбрали два черных, а из второго два белых
p_bb_ww = p1_0 * p2_2

# вероятность того, что из каждого ящика выбрали один белый и один черный
p_b_w = p1_1_1 * p2_1_1

print('Вероятность того, что ровно 2 мяча белые =', p_ww_bb + p_bb_ww + p_b_w)
```

```python
# вероятность того, что из каждого ящика выбрали два черных
p_bb_bb = p1_0 * p2_0

print('Вероятность того, что хотя бы один мяч белый =', 1 - p_bb_bb)
```
